```python
from PIL import Image
from IPython.core import display
from io import BytesIO

def display_pil_image(im):
    """displayhook function for PIL Images, rendered as PNG"""
    b = BytesIO()
    im.save(b, format='png')
    data = b.getvalue()

    ip_img = display.Image(data=data, format='png', embed=True)
    return ip_img._repr_png_()

def display_numpy_image(im, ax=None, cmap="Greys"):
    if ax == None:
        fig, ax = subplots(1,1)
    ax.imshow(im, cmap=cmap)
    ax.set_xticks([])
    ax.set_yticks([])   
    return ax

# register display func with PNG formatter:
png_formatter = get_ipython().display_formatter.formatters['image/png']
png_formatter.for_type(Image.Image, display_pil_image);
```

Load picture, convert to grayscale:

```python
image_rgb = array(img)
image_gray = 255-color.rgb2gray(image_rgb)
display_numpy_image(image_gray);
```

## Otsu segmentation

```python
thresh = filter.threshold_otsu(image_gray)
mask = image_gray >= thresh
display_numpy_image(mask);
```

## Edge detection

```python
edges = filter.canny(mask, sigma=1,
                     low_threshold=0.55, high_threshold=0.8)
display_numpy_image(edges);
```

```python
edges = filter.rank.maximum(edges, square(3))
display_numpy_image(edges);
```

## Working on a single seed

### Connected components

```python
img1_gray = color.rgb2gray(img1)
display_numpy_image(img1_gray);
```

```python
img2_gray = color.rgb2gray(img2)
display_numpy_image(img2_gray);
```

```python
th = img1_gray.mean()
print "Threshold:", th
mask1 = ~(img1_gray <= th)
display_numpy_image(mask1);
```

```python
label_im1, nb_labels1 = ndimage.label(mask1)
print nb_labels1 # how many regions?
display_numpy_image(label_im1, cmap=plt.cm.spectral)
sizes = ndimage.sum(mask1, label_im1, range(nb_labels1 + 1))
print "%.2g" % sum(sizes[1:])
```

### Blur mask

```python
display_numpy_image(img1);
```

```python
display_numpy_image(img2);
```

```python
th = img1_gray.mean()
print "Threshold:", th
mask1 = ~(img1_gray <= th)
display_numpy_image(mask1);
```

```python
th = img2_gray.mean()
print "Threshold:", th
mask2 = ~(img2_gray <= th)
display_numpy_image(mask2);
```

```python
blur1 = filter.rank.median(mask1, square(5))
display_numpy_image(blur1);
```

```python
blur2 = filter.rank.median(mask2, square(5))
display_numpy_image(blur2);
```

```python
edges1 = filter.canny(img1_gray, sigma=1)
edges1 = filter.rank.maximum(edges1, square(5))
display_numpy_image(edges1)
```

```python
edges2 = filter.canny(img2_gray, sigma=1)
edges2 = filter.rank.maximum(edges2, square(5))
display_numpy_image(edges2)
```

```python
mask_edges1 = edges1.astype(mask1.dtype) + mask1
display_numpy_image(mask_edges1)
```

```python
mask_edges2 = edges2.astype(mask2.dtype) + mask2
display_numpy_image(mask_edges2)
```

```python
filled1 = ndimage.morphology.binary_fill_holes(mask_edges1)
display_numpy_image(filled1)
```

```python
filled2 = ndimage.morphology.binary_fill_holes(mask_edges2)
display_numpy_image(filled2)
```

```python
print "Black: %.5f" % (sum(blur1 == 255)/float(sum(filled1)))
print "Black: %.5f" % (sum(blur2 == 255)/float(sum(filled2)))
```
