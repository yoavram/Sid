```python
from skimage import data, filter, color, measure
from scipy.ndimage import filters, label
from PIL import Image
from IPython.core import display
from io import BytesIO

def display_pil_image(im):
    """displayhook function for PIL Images, rendered as PNG"""
    b = BytesIO()
    im.save(b, format='png')
    data = b.getvalue()

    ip_img = display.Image(data=data, format='png', embed=True)
    return ip_img._repr_png_()

def display_numpy_image(im, ax=None, cmap="Greys", title=None):
    if ax == None:
        fig, ax = subplots(1,1)
    ax.imshow(im, cmap=cmap)
    ax.set_xticks([])
    ax.set_yticks([])
    if title:
        ax.set_title(title)
    return ax

# register display func with PNG formatter:
png_formatter = get_ipython().display_formatter.formatters['image/png']
png_formatter.for_type(Image.Image, display_pil_image);
```

```python
from skimage.morphology import dilation, erosion, square, disk
```

```python
def rgb2yuv(input): # modified from https://gist.github.com/bombless/4286560
    R, G, B = input[:,:,0],input[:,:,1],input[:,:,2]
    Y = array(0.299 * R + 0.587 * G + 0.114 * B, dtype=img.dtype)
    U = array(-0.147 * R + -0.289 * G + 0.436 * B, dtype=img.dtype)
    V = array(0.615 * R + -0.515 * G + -0.100 * B, dtype=img.dtype)
    output = input.copy()
    output[:,:,0],output[:,:,1],output[:,:,2] = Y,U,V
    return output
color.rgb2yuv = rgb2yuv
```

```python
img = Image.open("0615-1.jpg")
img = array(img)
h,w,_ = img.shape
display_numpy_image(img)
```

```python
hsv = color.rgb2hsv(img)
xyz = color.rgb2xyz(img)
yuv = color.rgb2yuv(img) # ycbcr
lab = color.rgb2lab(img)
hed = color.rgb2hed(img)
cie = color.rgb2rgbcie(img)
gray = color.rgb2gray(img)
```

```python
display_numpy_image(hsv, title='hsv')
display_numpy_image(xyz, title='xyz')
display_numpy_image(yuv, title='yuv')
display_numpy_image(lab, title='lab')
display_numpy_image(hed, title='hed')
display_numpy_image(cie, title='cie')
display_numpy_image(gray, title='gray')
```

```python
display_numpy_image(yuv[:,:,0], title='yuv R')
display_numpy_image(yuv[:,:,1], title='yuv G')
display_numpy_image(yuv[:,:,2], title='yuv B')
```

```python
def smooth(img):
    if len(img.shape) == 2:
        return filters.convolve(img, ones((10,10))/100, mode='mirror')
    elif len(img.shape) == 3:
        R, G, B = img[:,:,0],img[:,:,1],img[:,:,2]
        R, G, B = smooth(R), smooth(G), smooth(B)
        output = img.copy()
        output[:,:,0],output[:,:,1],output[:,:,2] = R, G, B
        return output       
    else:
        raise NotImplementedError()
```

```python
gray_smooth = smooth(gray)
display_numpy_image(gray_smooth, title="gray smooth")
yuv_smooth = smooth(yuv)
display_numpy_image(yuv_smooth, title="yuv smooth")
lab_smooth = smooth(lab)
display_numpy_image(lab_smooth, title="lab smooth")
```
